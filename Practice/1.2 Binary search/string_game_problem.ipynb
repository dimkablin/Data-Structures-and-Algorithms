## F. Игра со строкой

- ограничение по времени на тест: 2 секунды
- ограничение по памяти на тест: 512 мегабайт

У Пети есть слово $t$, он хочет, чтобы из него получилось слово $p$. Петя начинает вычеркивать буквы в некотором порядке, который задан перестановкой номеров букв слова $t$: $a_1, \ldots, a_{|t|}$. Заметим, что после вычеркивания буквы нумерация не меняется.

Его брат Вася боится, что Петя может удалить слишком много букв, и слово $p$ не получится. Задача Васи состоит в том, чтобы в некоторый момент времени остановить брата и закончить вычеркивание самому, получив после этого слово $p$. Так как Пете нравится это занятие, Вася хочет остановить его как можно позже. Ваша задача — сообщить, сколько букв может вычеркнуть Петя до того, как его остановит Вася.

Гарантируется, что слово $p$ можно получить вычеркиванием букв из $t$.

### Входные данные
Первая и вторая строки входного файла содержат слова $t$ и $p$, соответственно. Слова состоят из строчных букв латинского алфавита $(1 \leq |p| \leq |t| \leq 200\,000)$.

Следующая строка содержит перестановку $a_1, \ldots, a_{|t|}$ номеров букв, задающую порядок, в котором Петя вычеркивает буквы слова $t$ $(1 \leq a_i \leq |t|, все a_i \, различны)$.

### Выходные данные
Выведите одно число — максимальное число букв, которые может вычеркнуть Петя.

### Пример

| Входные данные | Выходные данные |
| --- | --- |
| abacbca \| abb \| 5 3 4 1 7 6 2 | 3 |
