## E. Уравнение

- ограничение по времени на тест: 2 секунды
- ограничение по памяти на тест: 256 мегабайт

Найдите такое число $x$, что $x^2 + \sqrt{x} = c$.

### Входные данные
В единственной строке содержится вещественное число $c$ $(1.0 \leq c \leq 10^{10})$.

### Выходные данные
Выведите одно число — искомый $x$. Ответ будет признан верным, если относительная или абсолютная погрешность не более $10^{-6}$.

### Примеры

| Входные данные | Выходные данные |
| --- | --- |
| 2.0 | 1.0 |
| 15.6 | 3.698232168829691 |

```python
from typing import Callable
import random


def test(solution: Callable):
    x = random.random() * 100 + 1
    c = x**2 + x**0.5

    print(f"C={c}, X={x}")
    x_pred = solution(c)

    assert abs(x_pred - x) < 10e-6, "Error"
    print(f"Predicted X={x_pred}")
```

```python
def solution(c):
    epsilon = 10e-7
    func = lambda x: x**2 + x**0.5
    l = 1
    r = c**3
    

    while l < r - epsilon:
        m = (l + r) / 2
        if func(m) < c:
            l = m
        else:
            r = m
    
    return (l+r)/2       
```

```python
solution(2)
```

```python
test(solution)
```
